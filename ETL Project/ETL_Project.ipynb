```python
flights_df = flights_df.rename(columns={"ORIGIN_AIRPORT": "airport_code", "CANCELLED": "CANCELLATION"})
flights_df.head()
```

```python
airports_df = airports_df.rename(columns={'IATA_CODE': "airport_code"})
airports_df.head()
```

```python
airports_df = airports_df[airports_df["airport_code"].isin(["ORD", "JFK", "SFO", "MCI", "ATL"])]
airports_df.head()
```

```python
merge_table = pd.merge(airports_df, flights_df, on="airport_code")
merge_table.head()
```

```python
merged_table = merge_table[merge_table["airport_code"].isin(["ORD", "JFK", "SFO", "MCI", "ATL"])]
merged_table.head()
```

```python
merged_table=merged_table[merged_table["DEPARTURE_DELAY"]>60.0]
```

```python
date=[]
for index, row in merged_table.iterrows():
    year= row["YEAR"]
    month= row["MONTH"]
    day= row["DAY"]
    time= row["SCHEDULED_DEPARTURE"]
    date.append(pd.to_datetime(f"{year}-{month}-{day}-{time}").round('60min'))
```

```python
merged_table['datetime'] = date
merged_table = merged_table.rename(columns={"CITY": "city",
                                            "DEPARTURE_DELAY": "departure_delay",
                                            "FLIGHT_NUMBER": "flight_number",
                                            "CANCELLATION": "cancellation"
                                                         })

merged_table.head()
```

```python
merged_table=merged_table.drop(['AIRPORT', 'STATE', 'YEAR', 'MONTH', 'DAY', 'SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'CANCELLATION_REASON'], axis=1)
merged_table=merged_table[['datetime', 'airport_code', 'city', 'flight_number', 'departure_delay', 'cancellation']]
```
