```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "mouse_metadata.csv"
study_results = "study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

mouse_metadata.head()
```

```python
study_results.head()
```

```python
#merge datasets 
merge_table = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")
merge_table
```

```python
#extract the drug reigmen, time point, and tumor volume columns
tumor_volume_df = merge_table.loc[:,['Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)']]
tumor_volume_df
```

```python
tumor_volume_df["Drug Regimen"].unique()
```

```python
summary_stats_table = tumor_volume_df.groupby(['Drug Regimen',]).agg({"Tumor Volume (mm3)" :["mean", "median", "var", "std", "sem",]})
summary_stats_table
```

```python
#create bar graph using pandas
data_points = merge_table['Drug Regimen'].value_counts()
data_points.plot(kind="bar")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Data Points")

plt.show()
```

```python
#create bar graph using pyplot
data_points = merge_table['Drug Regimen'].value_counts()
plt.bar(data_points.index.values,data_points.values)
plt.xlabel("Drug Regimen")
plt.xticks(rotation=90)
plt.ylabel("Number of Data Points")
plt.show()
```

```python
#create pie chart showing the distribution of male and female mice using pandas
counts = mouse_metadata.Sex.value_counts()
counts.plot(kind="pie",autopct='%2.2f%%')

plt.show()
```

```python
#create pie chart showing the distribution of male and female mice using pyplot
counts = mouse_metadata.Sex.value_counts()
plt.pie(counts.values,labels=counts.index.values,autopct='%2.2f%%')

plt.show()
```

```python
#create a line plot of timepoint vs tumor volume for a mouse treated with Capomulin 
capomulin_plot = merge_table.loc[merge_table['Drug Regimen'] == "Capomulin"]
mouse_info = capomulin_plot.loc[capomulin_plot['Mouse ID']== 'm601']
plt.plot(mouse_info['Timepoint'],mouse_info['Tumor Volume (mm3)'])
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Mouse m601 Capomulin treatment')
plt.show()
```

```python
#create a scatter plot of mouse weight vs average tumor volume using Capomulin
capomulin_plot = merge_table.loc[merge_table['Drug Regimen'] == "Capomulin"]
capomulin_avg = capomulin_plot.groupby(['Mouse ID']).mean()
plt.scatter(capomulin_avg['Weight (g)'],capomulin_avg['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

```python
#calculate the final tumor volume of each mouse across the 4 top treatment regimens.
final_volume = merge_table.groupby(["Mouse ID"]).max()
```

```python
final_volume = final_volume.reset_index()
merged_data = final_volume[['Mouse ID','Timepoint']].merge(merge_table,on=['Mouse ID','Timepoint'],how="left")
merged_data.head()
```

```python
#create a box plot of the final tumor volume of each mouse across four regimens of interest
capomulin = merged_data.loc[merged_data["Drug Regimen"] == "Capomulin"]['Tumor Volume (mm3)']
ramicane = merged_data.loc[merged_data["Drug Regimen"] == "Ramicane"]['Tumor Volume (mm3)']
infubinol = merged_data.loc[merged_data["Drug Regimen"] == "Infubinol"]['Tumor Volume (mm3)']
ceftamin = merged_data.loc[merged_data["Drug Regimen"] == "Ceftamin"]['Tumor Volume (mm3)']
```

```python
plt.boxplot([capomulin,ramicane,infubinol,ceftamin],labels=['Capomulin','Ramicane','Infubinol','Ceftamin'])
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
#find potential outliers and calculate IQR for capomulin
capomulin_quartiles = capomulin.quantile([.25,.5,.75])
capomulin_lowerq = capomulin_quartiles[0.25]
capomulin_upperq = capomulin_quartiles[0.75]
capomulin_iqr = capomulin_upperq-capomulin_lowerq
print(f"The iqr of capomulin is: {capomulin_iqr}")
```

```python
capomulin_lower_bound = capomulin_lowerq - (1.5*capomulin_iqr)
capomulin_upper_bound = capomulin_upperq + (1.5*capomulin_iqr)
print(f"Values below {capomulin_lower_bound} could be outliers.")
print(f"Values above {capomulin_upper_bound} could be outliers.")
```
