```python
import pandas as pd
import os
import numpy as np
import mpu
from geopy.distance import geodesic
```

```python
file1= "July_2019_citibike_data.csv"
file2= "January_2019_citibike_data.csv"
```

```python
df1 = pd.read_csv(file1)
df1.head()
```

```python
df2 = pd.read_csv(file2)
df2.head()
```

```python
#coordinates from start station and end stations
lat1=40.691966
lon1=-73.981302
lat2=40.693261
lon2=-73.968896
```

```python
#Using geopy library to calculate miles from Lat/ Long coordinates
import geopy.distance
from geopy.distance import vincenty

coords_1 = (lat1, lon1)
coords_2 = (lat2, lon2)

print(geopy.distance.distance(coords_1, coords_2).mi)
```

```python
new_df1= df1.rename(columns={'start station latitude': 'start_lats', 'start station longitude': 'start_longs', 'end station latitude': 'end_lats', 'end station longitude': 'end_longs', 'birth year': 'age'})
new_df2= df2.rename(columns={'start station latitude': 'start_lats', 'start station longitude': 'start_longs', 'end station latitude': 'end_lats', 'end station longitude': 'end_longs', 'birth year': 'age'})
```

```python
new_df1.dtypes
```

```python
#geopy distance in miles for start and end coordinates, adding new miles_traveled column
def distance(row):
    return geopy.distance.distance((row.start_lats, row.start_longs),(row.end_lats, row.end_longs)).mi
new_df1['miles_traveled']= new_df1.apply(lambda r: distance(r), axis=1)
```

```python
def distance(row):
    return geopy.distance.distance((row.start_lats, row.start_longs),(row.end_lats, row.end_longs)).mi
new_df2['miles_traveled']= new_df2.apply(lambda r: distance(r), axis=1)
```
