```python
#import dependencies
import matplotlib.pyplot as plt
import datetime as dt   
import pandas as pd
import numpy as np
import requests
import time
```

```python
#import API key
from api_keys import api_key
```

```python
from citipy import citipy
```

```python
#output file (csv)
output_data_file = "output_data/cities.csv"
```

```python
#range of latitudes and longitudes
lat_range = (-90, 90)
long_range = (-180, 180)
```

```python
#create lists
lat_longs = []
cities = []

#create a set of random lat and long combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1350)
longs = np.random.uniform(low=-180.000, high=180.000, size=1350)
lat_longs = zip(lats, longs)

#identify nearest city for each lat, long combination
for lat_long in lat_longs:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    
    #if the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

#print the city count
len(cities)
```

```python
#add the URL
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"

#build query
query_url = f"{url}units={units}&appid={api_key}&q="

#declare empty lists to store values for different cities
city_name =[]
city_lat = []
city_long = []
country = []
date = []
cloudiness = []
humidity = []
max_temp = []
wind_speed = []
```

```python
print("Beginning Data Retrieval")
print("------------------------")

#loop through data and get info
for city in cities:
    try:
         response = requests.get(query_url + city).json()
         #append the values in a list using loop
         city_name.append(response["name"])
         city_lat.append(response["coord"]["lat"])
         city_long.append(response["coord"]["lon"])
         country.append(response["sys"]["country"])
         date.append(response["dt"])
         humidity.append(response["main"]["humidity"])
         cloudiness.append(response["clouds"]["all"]) 
         max_temp.append(response["main"]["temp_max"])
         wind_speed.append(response["wind"]["speed"])
         #print the city and it's url along with count #
         print(f"Processing Record Number {cities.index(city)} as {city}")
         #using time.sleep to create time delay
         time.sleep(1)
              
    except :
              print("City not found. Skipping...")
              
print("-----------------------")
print("Data Retrieval Complete")
print("-----------------------")
```

```python
#create dataframe
weather_data = {
    "City":city_name,
    "Latitude":city_lat,
    "Longitude":city_long,
    "Country":country,
    "Date":date,
    "Humidity":humidity,
    "Cloudiness":cloudiness,
    "Max Temp":max_temp,
    "Wind Speed":wind_speed
}

weather_data =pd.DataFrame(weather_data)
```
